```python
# compute class weights
from sklearn.utils import class_weight
import numpy as np

"""
yaml
  0: discolor
  1: fly_die
  2: shift_sl
  3: flake
  4: scribe_line
  5: pad
  6: crack
  7: burr
  8: contamination
  9: chip
  10: particle
  11: corrosion
  12: scratch
"""

"""
scribe_line	pad     flake	burr	fly_die	contamination	crack	scratch	corrosion	chip	particle	shift_sl	discolor
1683	    5008	604	    601	    600	    605	            600	    610	    600	        631	    600	        600	        600

"""

class_count = {
    "discolor": 600,
    "fly_die": 600,
    "shift_sl": 600,
    "flake": 604,
    "scribe_line": 1683,
    "pad": 5008,
    "crack": 600,
    "burr": 601,
    "contamination": 605,
    "chip": 631,
    "particle": 600,
    "corrosion": 600,
    "scratch": 610
}

y = []
v = 0
for key, value in class_count.items():
    k = [key]*value
    y.extend(k)

classes = np.array(["discolor","fly_die","shift_sl","flake","scribe_line","pad","crack","burr","contamination","chip","particle","corrosion","scratch"])

cls_weights = class_weight.compute_class_weight("balanced", classes=classes, y=y)
cls_weights 
```

```python
from ultralytics import YOLO
```

```python
model = YOLO("yolov8m.pt")
model.train(
    data="final_crop/yolo_datasets/classes.yaml",
    epochs=300,
    patience=20,
    cls_weights=cls_weights.tolist(),
    device=0,
    project="YOLOv8",
    name="YOLOv8_crop_classweights",
    verbose=True
)
```

```python
model.val("final_crop/yolo_datasets/classes.yaml", split="test")
```
