```python
# compute class weights
from sklearn.utils import class_weight
import numpy as np

class_count = {
    "scribe_line": 1368, 
    "pad": 3562, 
    "flake": 73, 
    "burr": 288, 
    "fly_die": 80, 
    "contamination": 256, 
    "crack": 515, 
    "scratch": 410, 
    "corrosion": 54, 
    "chip": 616, 
    "particle": 169, 
    "shift_sl": 163, 
    "discolor": 41
}

y = []
v = 0
for key, value in class_count.items():
    k = [key]*value
    y.extend(k)

classes = np.array(["scribe_line", "pad", "flake", "burr", "fly_die", "contamination", "crack", "scratch", "corrosion", "chip", "particle", "shift_sl", "discolor"])

cls_weights = class_weight.compute_class_weight("balanced", classes=classes, y=y)
cls_weights 
```

```python
from ultralytics import YOLO
```

```python
model = YOLO("yolov8m.pt")
model.train(
    data="classes.yaml",
    epochs=300,
    patience=20,
    cls_weights=cls_weights.tolist(),
    device=0,
    project="YOLOv8m",
    name="YOLOv8m_classweights",
    verbose=True
)
```

```python
model.val(split="test")
```
